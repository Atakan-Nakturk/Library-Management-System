```python
class Library:

    def __init__(self):
        self.filename = "books.txt"
        try:
            self.file = open(self.filename, "a+")
            print(f"File '{self.filename}' opened successfully in 'a+' mode")

        except Exception as e:
            print(f"Error opening file '{self.filename}': {e}")

    def __del__(self):
        try:
            if self.file and not self.file.closed:
                self.file.close()
                print(f"File '{self.filename}' closed successfully")

        except Exception as e:
            print(f"Error closing file '{self.filename}': {e}")

    def list_books(self):
        try:
            self.file.seek(0)
            book_lines = self.file.read().splitlines()
            if book_lines:
              n = 0
              for line in book_lines:
                  title, author, _, _ = line.split(", ")
                  print(f"Book {n} - Title: '{title}', Author: '{author}'")
                  n+=1
            else:
              print("There aren't any books in the database")

        except Exception as e:
            print(f"Error listing books: {e}")

    def add_book(self):
        try:
            title = input("Enter the title of the book: ")
            author = input("Enter the author of the book: ")
            year = input("Enter the year of release: ")
            pages = input("Enter the number of pages: ")

            book_info = f"{title}, {author}, {year}, {pages}\n"
            self.file.write(book_info)
            print(f"Book '{title}' added successfully")

        except Exception as e:
            print(f"Error adding book: {e}")

    def remove_book(self):
        try:
            title_d = input("Enter the title of the book to remove: ")
            self.file.seek(0)

            book_lines = self.file.read().splitlines()
            matching_books = [line for line in book_lines if title_d in line]

            if not matching_books:
                print("No matching books found")
                return

            if len(matching_books) > 1:
                while True:
                    delete_all = input(f"{len(matching_books)} matching books found. Delete all? (y/n): ").lower()
                    if delete_all == "y":
                        break
                    elif delete_all == "n":
                        print("Provide additional information about the book you'd like to delete")
                        author_d = input("Enter the author of the book: ")
                        year_d = input("Enter the year of release: ")
                        pages_d = input("Enter the number of pages: ")

                        matching_books = [line for line in matching_books if title_d in line and author_d in line
                                          and year_d in line and pages_d in line]

                        if not matching_books:
                            print("No matching books found")
                            return

                        break

            updated_books = [line for line in book_lines if line not in matching_books]

            self.file.seek(0)
            self.file.truncate()

            for line in updated_books:
                self.file.write(line + "\n")

            print(f"{len(matching_books)} book(s) with specified attributes removed successfully")

        except Exception as e:
            print(f"Error removing book(s): {e}")
```

```python
lib = Library()

while True:
    print("\n*** MENU ***")
    print("1) List Books")
    print("2) Add Book")
    print("3) Remove Book")
    print("4) Exit")

    choice = input("Enter your choice (1-4): ")

    if choice == "1":
        lib.list_books()
    elif choice == "2":
        lib.add_book()
    elif choice == "3":
        lib.remove_book()
    elif choice == "4":
        del lib
        print("Exiting the program")
        break
    else:
        print("Invalid choice. Enter a number between 1-4")
```
